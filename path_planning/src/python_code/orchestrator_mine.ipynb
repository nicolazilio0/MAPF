```python
from scipy.optimize import minimize


robot_data = [
    {'center': np.array(shelfino0_path[0]), 'radius': 0.5, 'speed': 0.5},
    {'center': np.array(shelfino1_path[0]), 'radius': 0.5, 'speed': 0.5},
    {'center': np.array(shelfino2_path[0]), 'radius': 0.5, 'speed': 0.5}
]


# Example robot trajectories (replace with your actual data)
robot_trajectories = [shelfino0_path,shelfino1_path,shelfino2_path]  # Replace with actual trajectories

# Example collision pairs (replace with your actual data)
collision_pairs = [(0, 1), (0, 2), (1, 2)]


def calculate_total_distance(trajectory):
    """
    Calculate the total Euclidean distance traveled along the trajectory.
    
    Parameters:
    - trajectory: List of waypoints (coordinates) representing the robot's trajectory.
    
    Returns:
    - total_distance: Total Euclidean distance traveled along the trajectory.
    """
    total_distance = 0.0
    
    for i in range(1, len(trajectory)):
        # Euclidean distance between consecutive waypoints
        distance_between_points = np.linalg.norm(np.array(trajectory[i]) - np.array(trajectory[i - 1]))
        total_distance += distance_between_points
    
    return total_distance


# Calculate the total distance for each robot
total_distances = [calculate_total_distance(robot_trajectory) for robot_trajectory in robot_trajectories]

# Define the objective function to minimize (total time duration)
def objective_function(start_times):
    durations = [total_distances[i]*len(robot_trajectories[i]) for i in range(len(robot_trajectories))]
    print("durations:",durations)
    return np.sum(durations)

# Define collision avoidance constraints
def constraint_collision(start_times):
    constraints = []
    for i, j in collision_pairs:
        #for t in range(int(max(start_times[i], start_times[j])), int(max(start_times[i] + total_distances[i] / robot_data[i]['speed'], start_times[j] + total_distances[j] / robot_data[j]['speed'])) + 1):
        for t in range(int(min(start_times[i], start_times[j])), int(min(len(robot_trajectories[i]) ,  len(robot_trajectories[j])))):
            print("step:",t)

            contr_value=((robot_trajectories[i][t][0] - robot_trajectories[j][t][0])**2 - (robot_trajectories[i][t][1] - robot_trajectories[j][t][1])**2) < (robot_data[i]['radius'] + robot_data[j]['radius'])**2
            constr_to_app=max(0,contr_value)*100
            print("constr_to_app:",constr_to_app)
            constraints.append(constr_to_app)
    return constraints



def calculate_robot_trajectory(start_time, total_distance, robot_speed, original_trajectory):
    """
    Calculate the robot trajectory based on the original trajectory, adding a delay if needed.

    Parameters:
    - start_time: The start time of the trajectory.
    - total_distance: Total distance to be covered by the robot.
    - robot_speed: Speed of the robot.
    - original_trajectory: Original trajectory without any delay.

    Returns:
    - trajectory: List of waypoints representing the robot trajectory.
    """
    num_steps = len(original_trajectory)
    time_steps = np.linspace(start_time, start_time + total_distance / robot_speed, num_steps)
    
    # If start_time is greater than zero, add a delay to the trajectory
    delay = start_time * robot_speed
    trajectory = [(x + delay, y) for (x, y) in original_trajectory]
    
    return trajectory

# Initial guess for start times
initial_start_times = np.zeros(len(robot_data))

# Bounds for start times
bounds = [(0, None)] * len(initial_start_times)

# integer_constraint = {'type': 'eq', 'fun': lambda x: x % 1}

# Solve the optimization problem
result = minimize(objective_function, initial_start_times, constraints={'type': 'ineq', 'fun': constraint_collision}, bounds=bounds,options={'disp': True},method='trust-constr')
# Extract the optimal start times
optimal_start_times = result.x

print("Optimal Start Times:", optimal_start_times)


robot_trajectories = [calculate_robot_trajectory(start_time, total_distance, robot_data[i]['speed'], robot_trajectories[i])
                      for i, (start_time, total_distance) in enumerate(zip(optimal_start_times, total_distances))]

# Plot the original trajectories
plt.figure(figsize=(8, 8))
plt.scatter(obstacleList[..., 0], obstacleList[..., 1])
plt.scatter(goal[0], goal[1], color='red')



# Plot the optimized trajectories
for i, trajectory in enumerate(robot_trajectories):
    plt.plot([x for (x, y) in trajectory], [y for (x, y) in trajectory], label=f'Robot {i} (Optimized)', linestyle='--')

plt.legend()
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.title('Original and Optimized Robot Trajectories')
plt.show()



```
